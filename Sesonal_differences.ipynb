```python
ds_diff["pr"][0,0,0,6,:,:].plot()
```

```python
periods_ds = xr.DataArray(name = "period",
                          coords = {"year":ds_diff["year"]},
                          data   = comparions,
                          attrs  = {"long_name" : "comparison periods",
                                    "description" : "comparison periods"})

ds_diff = xr.merge([ds_diff,periods_ds])
```
